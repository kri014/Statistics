## Probablity Mass Function (PMF)

### rolling a single dice

```python
L=[]
for i in range(10000):
    L.append(random.randint(1,6))
    
```

```python
len(L)
```

```python
L[:5]
```

```python
s= (pd.Series(L).value_counts()/pd.Series(L).value_counts().sum()).sort_index()
```

```python
s
```

```python
s.plot(kind="bar")
```

#### Rolling two dice and getting sum  

```python
L=[]
for i in range(10000):
    a=random.randint(1,6)
    b=random.randint(1,6)
    L.append(a+b)
```

```python
L[:5]
```

```python
s= (pd.Series(L).value_counts()/pd.Series(L).value_counts().sum()).sort_index()
```

```python
s
```

```python
s.plot(kind="bar")
```

```python
#### Cumulative
```

```python
import numpy as np
```

```python
s_cum=np.cumsum(s)
```

```python
s_cum.plot(kind="bar")
```

## Parametric Density Estimation 

```python
import matplotlib.pyplot as plt
import numpy as np
from numpy.random import normal 
sample =normal(loc=50,scale=5,size=1000)
```

```python
sample
```

```python
sample.mean()
```

```python
# plot histogram to understand the distribution of data 
```

```python
plt.hist(sample,bins=10)
```

```python
#calculate sample mean and sample std dev
```

```python
sample_mean=sample.mean()
sample_std=sample.std()
```

```python
# fit the distribution with the above parameter 
from scipy.stats import norm
```

```python
dist=norm(sample_mean,sample_std)
```

```python
sample.min()
```

```python
sample.max()
```

```python
values = np.linspace(sample.min(),sample.max(),100)
```

```python
probablities = [dist.pdf(value) for value in values ]
```

```python
# plot the histogram and pdf 
plt.hist(sample,bins=10,density =True)
plt.plot(values,probablities)
```
